### 3 Анализ данных 

1. Просмотр информации о данных.
После загрузки данных в Pandas, просмотрим обзорную информацию о Дата фрейм с помощью  

###### .info(). 

Это даст нам преставление о количестве столбцов их название, типов дынных, количество заполненных строк. 

2. Просмотри первый и последний стобцы дата фрейма, 
###### .head(20)
###### .tail(2)

3. Проморим образец строки, с помощью метода 
###### .sample()

Зачем? .head() or .tail() может дать ложное представление о защищенности данных. 

4. Просмотрим количественные характеристики дата фрейма с помощью 
###### .describe() 

5. Проверяем количество пропусков.
###### isnull().sum()

6. Построим a boxplot. диаграмму для определения вбросов. 

7. Простроим heatmap.
Для понимания паттернов взаимосвязи, это даст нам возможность обнаружить взаимозависимый паттерн в большой корреляционной матрице.  

8. Построим другие графики для визуализации данных.

9. Очистим данные от вбросов, с помощью метода межквартального размаха. 

10. Определим количество пропусков и выберем способ обработки их, если количество пропусков достаточно большое количества в столбце, удалим столбец. В остальных случаях заполним пропуски средними значениями.
